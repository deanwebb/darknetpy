```python
from matplotlib import image, patches, pyplot as plt
from darknetpy.detector import Detector
```

```python
detector = Detector(
    '/Users/daniel/Workspace/darknet/cfg/coco.data', 
    '/Users/daniel/Workspace/darknet/cfg/yolov3.cfg', 
    '/Users/daniel/yolov3.weights'
)
```

```python
boxes = detector.detect('/Users/daniel/Workspace/darknet/data/dog.jpg')
```

```python
fig, ax = plt.subplots(1)
ax.imshow(image.imread('/Users/daniel/Workspace/darknet/data/dog.jpg'))

colors = ['r', 'b', 'y']

for i, box in enumerate(boxes):
    l = box['left']
    t = box['top']
    b = box['bottom']
    r = box['right']
    c = box['class']
    color = colors[i % len(colors)]
    
    rect = patches.Rectangle(
        (l, t), 
        r - l, 
        b - t,
        linewidth = 1, 
        edgecolor = color, 
        facecolor = 'none'
    )
    
    ax.text(l, t, c, fontsize = 12, bbox = {'facecolor': color, 'pad': 2, 'ec': color})
    ax.add_patch(rect)

plt.show()
```
